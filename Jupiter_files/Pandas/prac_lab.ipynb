```python
import numpy as np
import pandas as pd
```

```python
m = 1.5
c = 1
X = [10,20,30,33,44,55,60,70,90,99]
Y = []
```

```python
for i in range(len(X)):
    Y.append(m*X[i]+c)
print(Y)
```

```python
import matplotlib.pyplot as plt

# Sample data
# array1 = [1, 2, 3, 4, 5]
# array2 = [2, 4, 6, 8, 10]

# Plotting the data
plt.plot(X, label='X-values')
plt.plot(Y, label='Y-values')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Arrays Plot')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
