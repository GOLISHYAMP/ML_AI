```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
df = load_diabetes()
X = df.data
y = df.target
```

```python
X.shape
```

```python
X = X[:,:4]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y , test_size=0.2, random_state=1)
```

#### REMEMBER : Mostly fit_transform is done separately for X_train and y_train because the number of columns in them is different. So, once you used fit_transform for the train same can be used for test so there we will use transform only.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
y_train = sc.fit_transform(y_train.reshape(-1,1))
```

```python
y_train
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear', C=1.0, epsilon=0.5)
svr.fit(x_train, y_train)
```

```python
y_pred = svr.predict(x_test)
```

```python
y_pred = sc.inverse_transform(y_pred.reshape(-1,1))
```

```python
y_pred
```

```python
y_test
```

```python
y_test = y_test.reshape(-1,1)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
