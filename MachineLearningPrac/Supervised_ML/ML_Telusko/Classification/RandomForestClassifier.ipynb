```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../../../Datasets/Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
X = df.iloc[:, :2].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#### In Decision tree and random forest models there is no need of the scaling features.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, criterion='entropy')
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print("Accuracy of the Random forest classifier : ", accuracy_score(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

### Wow amazing results it got 95% accuracy. without any hyperparameter tuning and cross validations. Great!!!  Thats the power of ENSEMBLE LEARNING.
