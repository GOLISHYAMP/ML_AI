```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../Datasets/Position_Salaries.csv")
df.head()
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6), dpi = 300)
plt.scatter(x = x, y = y)
plt.xlabel('Experience')
plt.ylabel("salary")
plt.show()
```

### So the above graph show that we cannot use linear regression or multilinear regressor to fit the values. So, the best model to be use is polynomial regression.

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree= 4)
x_ploy = poly.fit_transform(x)
```

```python
x_ploy
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_ploy, y)
```

```python
plt.scatter(x, y , color = 'red')
plt.plot(x, lr.predict(poly.fit_transform(x)), color = 'blue')
plt.plot()
```
