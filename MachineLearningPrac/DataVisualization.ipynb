# implementation of figures

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0,10,11)
b = a**4

x = np.arange(0,10)
y = 2*x

fig = plt.figure(figsize=(5,5),dpi=500)
axes1 = fig.add_axes([0,0,0.5,0.5])
axes1.plot(a,b)
axes1.set_xlim(2,8)
axes1.set_ylim(2000,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')

axes1.set_title("A v/s B")



# fig = plt.figure()
axes1 = fig.add_axes([0.5,0.5,1,1])
axes1.plot(x,y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title("X v/s Y")


# fig.savefig(fname="mysample.png")
fig.savefig(fname="mysample1.png",bbox_inches = "tight")
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10)
y = x**4

a = np.linspace(0,10,11)
b = 2 * a

fig, axes = plt.subplots(nrows=1, ncols=2)
for axe in axes:
    axe.plot(x,y)
    axe.set_xlabel("X")
    axe.set_ylabel("Y")
    axe.set_title("X v/s Y")

# axes.plot(a,b)
# axes.set_xlabel("A")
# axes.set_ylabel("B")
# axes.set_title("A v/s B")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10)
y = x**4

a = np.linspace(0,10,11)
b = 2 * a

# fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (12,8))
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel("X")
axes[0][0].set_ylabel("Y")
axes[0][0].set_title("X v/s Y")

axes[1][1].plot(a,b)
axes[1][1].set_xlabel("A")
axes[1][1].set_ylabel("B")
axes[1][1].set_title("A v/s B")

fig.subplots_adjust(left=None, right=None,top=None, bottom=None, wspace=0.5, hspace=0.3)
```

## Lets make some styling
### It includes editing colors, legends, line width, markers

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,1)
y = x**2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'x v/s x')
ax.plot(x, y, label = 'x v/s x^2')
ax.legend(loc = 0)  # this will keep the legend at the best position

# ax.legend(loc = 1)  put the legend in the upper right corner
# ax.legend(loc = 2)  put the legend in the upper left corner
# ax.legend(loc = 3)  put the legend in the lower left corner
# ax.legend(loc = 4)  put the legend in the lower right corner

# we can also place the legend out of the plt by this
# ax.legend(loc = (1.1, 0.5))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, 'b--', label = 'x v/s x^2',alpha = 0.5)
ax.plot(x,x**3, 'g.-', label = 'x v/s x^3',alpha = 0.5)
ax.plot(x,x**4, 'r:', label = 'x v/s x^4', linewidth = 3)
ax.legend(loc = 0)
```

## Custom line design

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots()
line = axes.plot(x,x**3,label = 'x v/s x^3')
line[0].set_dashes((1,1,1,1,10,10))   # this is 1 unit of line , 1 unit of space, 1 unit of line, 1 unit of space, 10 unit of line and 10 unit of space.
```
