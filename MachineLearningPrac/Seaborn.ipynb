# * Distributions
### Basically distributive plots are used when we want to analyze any single feature.

```python
df.columns
```

### # Rugplot

```python
plt.figure(figsize=(12,8), dpi = 300)
sns.rugplot(x = 'Kms_Driven',data = df, height = 0.5)
# generally here in the rugplot we doesn't represent any data on y axis.
```

### # Hist and Dist plot
#### We do not come to know how many ticks marks are there per some segment in the rugplot, as many ticks are coinsiding with each other counting is not possibe. So to overcome this problem we can use Displot or Hist plot

```python
sns.displot(x = 'Kms_Driven', data = df, kde = True )
```

```python
sns.set_style(style='darkgrid')  # here you can give darkgrid, whitegrid, white, ticks
sns.histplot(x = 'Kms_Driven', data = df, kde= True, bins = 50, color = 'red',
              edgecolor = 'black', lw = 10, ls = '--')
```

## # Kde plot (Kernel density estimation)

```python
sns.kdeplot(x = 'Kms_Driven', data = df, clip = [0, 200000] , 
            color = 'red', fill=True, lw = 2 , bw_adjust=0.5)
```

#### Well till now in the distribution plots we have worked only on a single feature
#### but there is a 2 dimensional kde plot which can represent two features

```python
df.columns
```

```python
sns.kdeplot(data = df, x = 'Selling_Price', y = 'Present_Price', fill = True, 
            color = 'red',clip=[-10,40] )
```

## # Barplot

#### in countplot we saw that y-axis is only used for counting but by using barplot we can expand our visualization by specifying a specific continuous feature for the y-axis

```python
df.columns
```

```python
plt.figure(figsize=(12,8), dpi = 300)
sns.barplot(x = 'Fuel_Type', y = 'Present_Price', data = df, 
            hue = 'Transmission',estimator=np.mean, errorbar='sd')
```

## # Boxplot

```python
df = pd.read_csv('Datasets\car data.csv')
```

```python
df.columns
```

```python
plt.figure(figsize=(12,8),dpi = 300)
sns.boxplot(x = 'Fuel_Type', y = 'Present_Price', data = df,
             hue = 'Transmission')
plt.legend(bbox_to_anchor = (1.05,0.5))
# plt.legend(loc = 0)
```

### # Voilinplot

```python
plt.figure(figsize=(12,8) , dpi = 300)
sns.violinplot(x = 'Present_Price', y = 'Fuel_Type', data = df,
                hue = "Transmission")
```

```python
plt.figure(figsize=(12,8) , dpi = 300)
sns.violinplot(x = 'Present_Price', y = 'Fuel_Type', data = df, hue = "Transmission",
               split=True, inner = 'quartiles')
```

### # Swarmplot

```python
plt.figure(figsize=(12,8),dpi = 300)
sns.swarmplot(x = 'Present_Price', y = 'Fuel_Type', data = df,
               hue = 'Transmission')
```

### # Boxenplot

```python
plt.figure(figsize=(12,8), dpi= 300)
sns.boxenplot(x = 'Present_Price', y = 'Fuel_Type', data = df,
               hue = 'Transmission')
```

# * Comparison Plots

### # Joint Plot

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.jointplot(x = 'Present_Price', y = 'Selling_Price', data = df)
```

#### here we can change the style of graph with the help of kind = ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'], by default it is set to scatterplot 

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.jointplot(x = 'Present_Price', y = 'Selling_Price', data = df, kind = 'kde')
```

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.jointplot(x = 'Present_Price', y = 'Selling_Price', data = df, kind = 'hex')
```

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.jointplot(x = 'Present_Price', y = 'Selling_Price', data = df, kind = 'reg')
```

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.jointplot(x = 'Present_Price', y = 'Selling_Price', data = df, kind = 'resid')
```

### # Pairplot
#### all it needs the dataset, rest of all it do on itself. All the numeric data info will be displayed in graphical manner

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.pairplot(data = df)
```

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.pairplot(data = df, hue = "Fuel_Type")
```

```python
plt.figure(figsize = (12,8), dpi = 300)
sns.pairplot(data = df, hue = "Fuel_Type", corner= True)
```

## * Seaborn Grid

### # Cat plot

```python
plt.figure(figsize=(12, 8), dpi = 300)
sns.catplot(x = 'Fuel_Type', y = 'Present_Price', data = df,
             kind= 'bar')
# kind can be : ['point', 'bar', 'strip', 'box', 'swarm', 'voilin', 'boxen]
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 8), dpi = 300)
sns.catplot(x = 'Fuel_Type', y = 'Present_Price', data = df, kind= 'bar', 
            row = 'Transmission', col = 'Owner')
```

## # Pairgrid

```python
plt.figure(figsize=(12,8), dpi = 300)
g = sns.PairGrid(data = df, hue = 'Fuel_Type', hue_kws={'marker' : ['O','+']} )
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot, lw = 2)
g = g.map_lower(sns.kdeplot, colors = 'red')
```

## # FacetGrid

```python
g = sns.FacetGrid(data= df, col = 'Transmission', row = 'Fuel_Type')
g = g.map(plt.scatter, 'Present_Price', 'Selling_Price')
g.add_legend(loc = 0)
plt.subplots_adjust(hspace=0.4, w_space = 0.4)
```
